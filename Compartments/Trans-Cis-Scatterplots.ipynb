```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
import math
```

```python
names_hTERT = [
    'hTERT-TP2-R1',
    'hTERT-TP4-R1-1',
    'hTERT-TP4-R1-2',
    'hTERT-TP5-R1',
    'hTERT-TP6-R1'
]

names_PDL = [
    'PDL25-TP2-R1',
    'PDL33-TP4-R1',
    'PDL37-TP5-R1',
    'PDL46-TP6-R1'
]
```

```python
d1 = pd.read_csv('Perpendicular_distance/Original.tsv',sep='\t')
d2 = pd.read_csv('Trans-analysis/Original.tsv',sep='\t')
```

```python
d2.head()
```

```python
fig = plt.figure(figsize=(40, 8))
gs = gridspec.GridSpec(nrows=1, ncols=5)

for i in range(5):
    df=d1[['chrom','start','end',names_hTERT[i]]]
    df['trans']=d2[[names_hTERT[i]]]
    df=df.rename(columns={names_hTERT[i]:"cis"}).dropna()
    #df.head()
    ax0 = fig.add_subplot(gs[0,i])
    sns.set()
    sns.scatterplot(x="cis",y="trans",data=df)
    plt.title(names_hTERT[i],fontdict = {'fontsize' : 30})
```

```python
fig = plt.figure(figsize=(40, 10))
gs = gridspec.GridSpec(nrows=1, ncols=4)

for i in range(4):
    df=d1[['chrom','start','end',names_PDL[i]]]
    df['trans']=d2[[names_PDL[i]]]
    df=df.rename(columns={names_PDL[i]:"cis"}).dropna()
    #df.head()
    ax0 = fig.add_subplot(gs[0,i])
    sns.set()
    sns.scatterplot(x="cis",y="trans",data=df)
    plt.title(names_PDL[i],fontdict = {'fontsize' : 30})
```
