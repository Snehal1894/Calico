```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
files_hTERT =[
    'MC-HiC-DpnII-WI38-hTERT-TP2-R1.cis.vecs.tsv',
    'MC-HiC-DpnII-WI38-hTERT-TP4-R1-1.cis.vecs.tsv',
    'MC-HiC-DpnII-WI38-hTERT-TP4-R1-2.cis.vecs.tsv',
    'MC-HiC-DpnII-WI38-hTERT-TP5-R1.cis.vecs.tsv',
    'MC-HiC-DpnII-WI38-hTERT-TP6-R1.cis.vecs.tsv'
]

files_PDL =[
    'MC-HiC-DpnII-WI38-PDL25-TP2-R1.cis.vecs.tsv',
    'MC-HiC-DpnII-WI38-PDL33-TP4-R1.cis.vecs.tsv',
    'MC-HiC-DpnII-WI38-PDL37-TP5-R1.cis.vecs.tsv',
    'MC-HiC-DpnII-WI38-PDL46-TP6-R1.cis.vecs.tsv'
]

names_hTERT = [
    'hTERT-TP2-R1',
    'hTERT-TP4-R1-1',
    'hTERT-TP4-R1-2',
    'hTERT-TP5-R1',
    'hTERT-TP6-R1'
]

names_PDL = [
    'PDL25-TP2-R1',
    'PDL33-TP4-R1',
    'PDL37-TP5-R1',
    'PDL46-TP6-R1'
]
```

```python
fig = plt.figure(figsize=(35, 35))
gs = gridspec.GridSpec(nrows=5, ncols=5)
fig.suptitle('Eigen Vector 2 hTERT samples',size=30)
for i in range(5):
    f1 = pd.read_csv(files_hTERT[i], sep='\t', header=0)
    for j in range(i,5):
        f2 = pd.read_csv(files_hTERT[j], sep='\t', header=0)
        f3=f1[['chrom','start','end','E2']]
        f3[['E2-new']]=f2[['E2']]
        f3.dropna()
        ax0 = fig.add_subplot(gs[i, j])
        ax0.scatter(f3['E2'],f3['E2-new'])
        ax0.set_xlabel(names_hTERT[i])
        ax0.set_ylabel(names_hTERT[j])
```

```python
fig = plt.figure(figsize=(35, 35))
gs = gridspec.GridSpec(nrows=4, ncols=4)
fig.suptitle('Eigen Vector 2 PDL samples',size=30)
for i in range(4):
    f1 = pd.read_csv(files_PDL[i], sep='\t', header=0)
    for j in range(i,4):
        f2 = pd.read_csv(files_PDL[j], sep='\t', header=0)
        f3=f1[['chrom','start','end','E2']]
        f3[['E2-new']]=f2[['E2']]
        f3.dropna()
        ax0 = fig.add_subplot(gs[i, j])
        ax0.scatter(f3['E2'],f3['E2-new'])
        ax0.set_xlabel(names_PDL[i])
        ax0.set_ylabel(names_PDL[j])
```
